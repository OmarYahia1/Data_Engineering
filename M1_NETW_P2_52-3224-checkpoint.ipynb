```python
#The first step categorizes 'annual income' into discrete ranges. This is done by dividing the income data into specific intervals, 
#such as "Very-Low" for incomes up to $50,000, "Low" for incomes between $50,001 and $100,000, and so on, up to "High" for incomes above $250,000. 
#This new categorization creates a new column in the DataFrame that assigns each income value to its corresponding income range.
df['income_range'] = pd.cut(df['annual_inc'], bins=[0, 50000, 100000, 150000, 200000, 250000, np.inf],
                            labels=['Very-Low', 'Low', 'Mid-Low', 'Mid', 'Mid-High','High'])


# The next step involves grouping the data by the newly created income ranges to determine the most common job title for each range. 
#This is achieved by calculating the mode, which is the most frequently occurring value in the employment title column for each income category. 
mode_annual_inc = df.groupby('income_range')['emp_title'].agg(lambda x: x.mode()[0] if not x.mode().empty else "Unknown")

# After determining the most common job title for each income range, the code fills in missing values in the emp_title column. 
#It replaces any missing values with the corresponding most common job title from the same income category.
df['emp_title'] = df['emp_title'].fillna(df['income_range'].map(mode_annual_inc))

# A similar approach is then used to determine the most common employment length for each income range. The data is grouped by income ranges again, 
#and the mode of the employment length is calculated for each group.
mode_annual_inc2 = df.groupby('income_range')['emp_length'].agg(lambda x: x.mode()[0] if not x.mode().empty else "Unknown")

# Any missing values in the employment length column are filled using the most common values determined for each income category, 
#similar to the process used for emp_title.
df['emp_length'] = df['emp_length'].fillna(df['income_range'].map(mode_annual_inc2))

# Mapping ategorical values in the 'emp_length' column  into numerical values. 
#This transformation is useful, where numerical data is typically required.

df = dropcol(df,'income_range')
df.emp_title.head(8)
```

```python
#The Label Encoding technique converts categorical variables into numerical values, allowing models to interpret ordinal relationships between 
#categories, such as severity or ranking, which can enhance the model's predictive power.

# This allows us to create a hierarchical representation of loan status,
# where 'Fully Paid' has the highest value (6) and 'Default' has the lowest (0).
loan_status_order = {
    'Fully Paid': 6,
    'Current': 5,
    'In Grace Period': 4,
    'Late (16-30 days)': 3,
    'Late (31-120 days)': 2,
    'Charged Off': 1,
    'Default':0
}


# rank the grades from 'A' to 'G' based on their performance, allowing for easier comparison.
letter_grade_mapping = {
    'A': 6,
    'B': 5,
    'C': 4,
    'D': 3,
    'E': 2,
    'F': 1,
    'G': 0
}
# Map verification status to numerical values to capture the confidence level of the data.
verification_status_mapping = {
    'Verified': 2,
    'Source Verified': 1,
    'Not Verified': 0
}
# Map home ownership status to numerical values to quantify the stability of the borrower's 
# financial situation. This can be a significant feature for predicting loan repayment.
home_ownership_mapping = {
    'OWN': 4,
    'MORTGAGE': 3,
    'RENT': 2,
    'ANY': 1,
    'NONE': 0
}
# Function to apply the mapping to the specified column in the DataFrame.
def labelEnc(df,col,mapping):
    df[col] = df[col].map(mapping)
    return df
# Map these features to numerical values
df = labelEnc(df, 'loan_status', loan_status_order)
df = labelEnc(df, 'letter_grade', letter_grade_mapping)
df = labelEnc(df, 'verification_status', verification_status_mapping)
df= labelEnc(df, 'home_ownership', home_ownership_mapping)


df.verification_status.head(), df.letter_grade.head(), df.loan_status.head(), df.home_ownership.head()
```

```python
#The code compares two DataFramesone original and one modifiedto track changes and creates a "lookup table" that records any differences between them,
#particularly highlighting imputed values. It includes missing values that have been filled, and encoded values.
def map_grade(value):
    if 1 <= value <= 5:
        return 'A'
    elif 6 <= value <= 10:
        return 'B'
    elif 11 <= value <= 15:
        return 'C'
    elif 16 <= value <= 20:
        return 'D'
    elif 21 <= value <= 25:
        return 'E'
    elif 26 <= value <= 30:
        return 'F'
    elif 31 <= value <= 35:
        return 'G'
    else:
        return value 

def lookup(df, original_df):
    # Initialize an empty list to store the lookup table records
    lookup_records = []
    # Create a dictionary to track seen original values for each column
    seen_values = {column: set() for column in original_df.columns}

    # Iterate over the columns to compare the original and modified data
    for column in original_df.columns:
        for original_value, modified_value in zip(original_df[column], df[column]):
            # If the original and modified values are different
            if original_value != modified_value:
                if pd.isnull(original_value):
                    original_value_display = "missing"  # Set to "missing" if original value is NaN
                else:
                    original_value_display = original_value

                # Handle special case for 'int_rate'
                if column == 'int_rate' and original_value_display == "missing":
                    modified_value = f"{modified_value} (median)"
                    
                if (column == 'emp_length' or column == 'emp_title' )and original_value_display == "missing":
                    modified_value = f"{modified_value} (mode)"
                # Check if the original value has been seen before
                if original_value_display not in seen_values[column]:
                    lookup_records.append({
                        'Column': column,
                        'Original': original_value_display,
                        'Imputed': modified_value
                    })
                    seen_values[column].add(original_value_display)  # Mark this original value as seen
            
            # Handle special case for 'grade'
            if column == 'grade':
                original_value_display = original_value
                modified_value = map_grade(modified_value)

                # Check if the original value has been seen before
                if original_value_display not in seen_values[column]:
                    lookup_records.append({
                        'Column': column,
                        'Original': original_value_display,
                        'Imputed': modified_value
                    })
                    seen_values[column].add(original_value_display)  # Mark this original value as seen

    # Create a DataFrame for the lookup table
    lookup_table_df = pd.DataFrame(lookup_records)
    return lookup_table_df

# Call the lookup function.
lookup_table_df = lookup(df, original_df)

# Save the lookup table to a CSV file.
lookup_table_df.to_csv('fintech_data_NETW_P2_52_3224_lookup_table.csv', index=False)

lookup_table_df.tail()
```

```python
#Capping and flooring techniques are used to manage extreme values (outliers) in a dataset by reducing skewness, making the distribution more
#balanced. Flooring replaces any value below a certain point (usually 10th quantile) with that point, while capping replaces values above a set point 
#(usually 90th quantile). This helps reduce the impact of extreme values, making data analysis more reliable. Overall, these techniques improve the 
#accuracy of analyses in fields like finance and customer behavior, where outliers can distort results.

#a method to handle outliers using Quantile-based Flooring and Capping.
def handle_outliers(df, col):
    floor = df[col].quantile(0.10)
    cap = df[col].quantile(0.90)
    print("cap: ",cap)
    print("floor: ",floor)
    df[col] = np.where(df[col] <floor, floor,df[col])
    df[col] = np.where(df[col] >cap,cap,df[col])
    return df
```

```python
#Calculate the monthly installment (M = P * r * (1 + r)^n / ((1 + r)^n - 1))
def calculate_monthly_installment(loan_amount, annual_int_rate, term_in_months):
    # Convert annual interest rate to monthly by dividing by 12.
    r = annual_int_rate / 12
    n = term_in_months  # Number of months (loan term).
    if r == 0:  # If the interest rate is zero, return the simple installment calculation (loan amount divided by term).
        return loan_amount / n
    # M = P * r * (1 + r)^n / ((1 + r)^n - 1)
    return loan_amount * r * (1 + r)*n / ((1 + r)*n - 1)

# Extract the numeric part from the 'term' column (e.g., "36 months" -> 36).
df['term_int'] = df.term.str.extract('(\d+)').astype(int)

# Calculate the monthly installment
df['installment_per_month'] = df.apply(
    lambda row: calculate_monthly_installment(row['loan_amount'], row['int_rate'], row['term_int']),
    axis=1
)
# Drop the temporary 'term_int' column after use, as it's no longer needed.
df = dropcol(df, 'term_int')
df.installment_per_month.head()
```

```python
# One-hot encode categorical features that have no inherent order, such as 'term', 'type', and 'purpose'.
# This technique creates binary columns for each category, allowing the model to treat them independently
# while avoiding the introduction of any ordinal relationships that dont exist in the original data.
#Instead of doing one hot encoding on all the values in each feature and the dataset will be very large, I will take the topmost 3 most 
#frequent values in each feature.


# Function that calculates the top x most frequent values in a categorical feature.
def calculate_top_categories(df, variable, how_many):
    return [
        x for x in df[variable].value_counts().sort_values(
            ascending=False).head(how_many).index
    ]

# Manually encode the most frequent values in a categorical feature.
def one_hot_encode(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable + '_' + label] = np.where(
            df[variable] == label, 1, 0)

# List of categorical columns to be one-hot encoded
categorical_columns = ['emp_title', 'emp_length', 
                       'zip_code', 'state', 'term', 
                       'type', 'purpose']

# Loop through each categorical column and apply one-hot encoding
for col in categorical_columns:
    top_x = calculate_top_categories(df, col, 3)
    one_hot_encode(df, col, top_x)
    
df.columns
```
